```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
font_size = 25
```

```python
def get_blink_bw(path):
    data_points = []
    bw_points = []
    nccl_bw_points = []
    time_points = []
    with open(path, 'r') as f:
        lines = f.readlines()
        for l in lines:
            if "Total Average" in l:
                str_lst = l.split(" ")
                data_points.append(int(str_lst[7]))
                bw_points.append(float(str_lst[-2]))
                
    return np.array(data_points), np.array(bw_points)
```

```python
def get_baseline_bw(path):
    with open(path, 'r') as f:
        lines = f.readlines()
        bw_points = []
        for l in lines:
            if "Average" in l:
                str_lst = l.split(" ")
                bw_points.append(float(str_lst[-2]))

    return np.array(bw_points)
```

# Preliminary Best Setting and Results

## Broadcast

```python
blinkplus_broad_3group_path = "../test_bandwidth/logs/broadcast_user01_optimal.log"
nccl_broad_path = "../test_bandwidth/logs/broadcast_user01_nccl_baseline.log"
blink_data_points, blink_bw_points = get_blink_bw(blinkplus_broad_3group_path)
nccl_bw_points = get_baseline_bw(nccl_broad_path)

fig = plt.figure(figsize = (12, 5))
X_axis = np.arange(len(data_points))
data_points_ticks = [str(item) for item in data_points]
plt.bar(X_axis - 0.15, nccl_bw_points, width = 0.3, label='NCCL')
plt.bar(X_axis + 0.15, blink_bw_points, width = 0.3, label='Blink+')
plt.xticks(X_axis, data_points_ticks)

plt.xlabel("Data (MB)", fontsize=font_size-5)
plt.ylabel("BW (GB/s)", fontsize=font_size-5)
plt.legend(fontsize=font_size-5)
plt.xticks(fontsize=font_size-5)
plt.yticks(fontsize=font_size-5)
plt.title(r"GPU 0-1 BroadCast Benchmark ($\bf{2}$ Side Channels)" + "\n" + r"$\bf{Theoretical \; Max \;BW = 60GB/s}$", fontsize=font_size)
plt.savefig("figs/gpu0_1_broadcast_optimal.png", bbox_inches = 'tight')


blinkplus_broad_3group_path = "../test_bandwidth/logs/broadcast_user03_optimal.log"
nccl_broad_path = "../test_bandwidth/logs/broadcast_user03_nccl_baseline.log"
blink_data_points, blink_bw_points = get_blink_bw(blinkplus_broad_3group_path)
nccl_bw_points = get_baseline_bw(nccl_broad_path)

fig = plt.figure(figsize = (12, 5))
X_axis = np.arange(len(data_points))
data_points_ticks = [str(item) for item in data_points]
plt.bar(X_axis - 0.15, nccl_bw_points, width = 0.3, label='NCCL')
plt.bar(X_axis + 0.15, blink_bw_points, width = 0.3, label='Blink+')
plt.xticks(X_axis, data_points_ticks)

plt.xlabel("Data (MB)", fontsize=font_size-5)
plt.ylabel("BW (GB/s)", fontsize=font_size-5)
plt.legend(fontsize=font_size-5)
plt.xticks(fontsize=font_size-5)
plt.yticks(fontsize=font_size-5)
plt.title(r"GPU 0-1 BroadCast Benchmark ($\bf{2}$ Side Channels)" + "\n" + r"$\bf{Theoretical \; Max \;BW = 60GB/s}$", fontsize=font_size)
plt.savefig("figs/gpu0_3_broadcast_optimal.png", bbox_inches = 'tight')
```
